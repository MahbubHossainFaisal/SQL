#### My Solution:

    with cte as
    (
        Select *, Min(order_type) Over(partition by Customer_id) as min_order_type
        from Orders
    )

    Select order_id,customer_id, order_type
    where order_type = min_order_type
    order by order_type Desc;

#### Tutorial
    
    with cte as
    (
        Select *, Min(order_type) Over(partition by Customer_id) as min_order_type
        from Orders
    )

    Select order_id,customer_id, order_type
    where (order_type + min_order_type <> 1)
    order by order_type Desc;
