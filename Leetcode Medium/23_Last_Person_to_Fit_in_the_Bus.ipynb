    WITH cte AS
    (
        SELECT 
            person_id,
            person_name,
            weight,
            SUM(weight) OVER (ORDER BY turn ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS total_weight,

            turn
            From Queue
            Order by turn
    )

    SELECT person_name FROM cte
    where total_weight <=1000
    Order by turn DESC limit 1;
