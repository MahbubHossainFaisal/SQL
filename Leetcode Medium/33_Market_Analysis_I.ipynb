    with cte as
    (
        Select u.user_id,u.join_date,o.buyer_id,o.order_date,count(*) as number
        from Users u LEFT JOIN Orders o on u.user_id = o.buyer_id
        where year(o.order_date) = '2019'
        group by o.buyer_id
    )

    select us.user_id as buyer_id,us.join_date,coalesce(c.number,0) as orders_in_2019
    from Users us LEFT JOIN cte c on us.user_id = c.user_id
    group by us.user_id
