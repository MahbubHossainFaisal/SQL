    with cte as
    (
        select o.product_id,o.order_id,o.order_date,max(o.order_date) over(partition by o.product_id) as latest_order_date
        ,p.product_name
        from Orders o JOIN Products p on o.product_id = p.product_id
    )

    select product_name,product_id,order_id,order_date from cte where order_date = latest_order_date
    ORDER BY product_name, product_id, order_id;
