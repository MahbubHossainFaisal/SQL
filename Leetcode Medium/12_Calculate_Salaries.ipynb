    with cte as
    (
        Select company_id,employee_id,employee_name,
        max(salary) Over(Partition by company_id) as max_salary,
        (Case when max_salary between 1000 and 10000 then round(salary-salary*0.24)
             when max_salary > 10000 then (salary-salary*0.49)) as salary_after_tax
    )

    select company_id,employee_id, salary_after_tax as salary
    from cte
    Order by company_id
