```python
    with cte as
    (
        select c.customer_name,o.customer_id,o.order_id,o.order_date,o.cost,
        RowNumber() over(partition by customer_id order by order_date DESC) as o.order_number
        from customer c JOIN Order o on c.customer_id = o.order_id
    )

    select customer_name,customer_id,order_id,order_date from cte
    where order_number <=3
    order by customer_name
```
